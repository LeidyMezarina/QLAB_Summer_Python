```python
import pandas as pd

class DataFrameAnalyzer:  
    def __init__(self, dataframe):   #atributtes
        self.dataframe = dataframe

    def descriptive(self, window_size=None):
        # Select columns from index 8 to 32
        selected_columns = self.dataframe.iloc[:, 8:33]

        # Convert selected columns to numeric, coercing errors to NaN
        selected_columns_numeric = selected_columns.apply(pd.to_numeric, errors='coerce')

        # Calculate statistics along rows
        row_means = selected_columns_numeric.mean(axis=1)
        row_max = selected_columns_numeric.max(axis=1)
        row_min = selected_columns_numeric.min(axis=1)
        row_std = selected_columns_numeric.std(axis=1)

        # Create a new DataFrame with the results
        result_df = pd.DataFrame({
            'Mean': row_means,
            'Max': row_max,
            'Min': row_min,
            'Std': row_std
        })

        # Calculate mean of the years
        column_means = selected_columns_numeric.mean(axis=0)

        # Find the column with the best mean
        best_mean_column = column_means.idxmax()
        worst_mean_column = column_means.idxmin()

        # Display the results
        print("\nDescriptive Statistics (Row-wise):")
        print(result_df)

        print("\nMean of each column:")
        print(column_means)

        print("\nColumn with the best mean:", best_mean_column)
        print("Column with the worst mean:", worst_mean_column)

        if window_size is not None:
            # Calculate rolling average using a window of specified size
            rolling_avg = selected_columns_numeric.rolling(window=window_size, axis=1).mean()
            print("\nRolling Average:")
            print(rolling_avg)

file_path = r"C:\Users\Hp\Documents\GitHub\QLAB_Summer_Python\_data\bogota_firms.xlsx"
df_Results = pd.read_excel(file_path, sheet_name="Results")

# Instantiate the class and perform descriptive statistics
analyzer = DataFrameAnalyzer(df_Results)
analyzer.descriptive(window_size=3)
```
