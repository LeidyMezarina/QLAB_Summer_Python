```python
import pandas as pd
import statistics

# Define the class ListStats
class ListStats:
    # Add the __init__ method
    def __init__(self, numbers):
        # Assign the argument to the attribute self.numbers
        self.numbers = numbers
    
    # Define the descriptive method
    def descriptive(self):
        # Calculate the mean, max, min, median, and rolling average of the numbers
        mean = statistics.mean(self.numbers)
        max = max(self.numbers)
        min = min(self.numbers)
        median = statistics.median(self.numbers)
        rolling_average = sum(self.numbers) / len(self.numbers)
        # Return a dictionary with the results
        return {"mean": mean, "max": max, "min": min, "median": median, "rolling_average": rolling_average}

# Read the Excel file
file_path = r"C:\Users\Hp\Documents\GitHub\QLAB_Summer_Python\_data\bogota_firms.xlsx"
df_Results = pd.read_excel(file_path, sheet_name="Results")

# Select the column "Operating revenue (Turnover)th LCU 2023"
columna_1 = df_Results["Operating revenue (Turnover)th LCU 2023"]

# Convert the column to a list of numbers
columna_1 = [float(x) for x in columna_1]

# Create an object of the class ListStats with the column as an argument
my_stats = ListStats(columna_1)

df_Results = pd.read_excel(file_path, sheet_name="Results", header=0)

# Print the result of the descriptive method in a readable format
print(f"The descriptive statistics of the column are: {my_stats.descriptive()}")
```
